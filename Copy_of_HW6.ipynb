```python
#Your code goes here.
#In the same Python notebook, read WorldCupMatches.csv by using Pandas.
world_cup = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DATA/WorldCupMatches.csv')
world_cup.head()
```

```python
#Then using Pandas groupby feature on ’Home Team Initials’ and ’Away Team Initials’ find how many times these countries’ national teams played a match
#You can use the count() after grouping. These counts will be the weight attribute when you create your final data frame explained below.
team_count = world_cup.groupby(['Home Team Initials', 'Away Team Initials']).count().reset_index()
team_count = team_count[['Home Team Initials', 'Away Team Initials', 'Year']]
team_count.columns = ['Home Team Initials', 'Away Team Initials', 'weight']
team_count.head()
```

```python
#Find the total goals scored by each home team for the matches played above by using groupby as earlier and sum on the ’Home Team Goals’ column.
#These summations will be your HomeGoalTotal’ attribute below.
home_goals = world_cup.groupby('Home Team Initials')['Home Team Goals'].sum().reset_index()
home_goals.columns = ['Home Team Initials', 'HomeGoalTotal']
home_goals.head()
```

```python
#Use the values you found above, come up with a new data frame called team_pairs which has ’Home Team Initials’, ’Away Team Initials’, ’weight’, ’HomeGoalTotal’ columns where latter two columns should have the values from your earlier steps.
team_pairs = pd.merge(team_count, home_goals, left_on='Home Team Initials', right_on='Home Team Initials')
team_pairs.head()
```

```python
#Write your team_pairs to a csv file by using Pandas to_csv feature. (Please ignore the fact that there may be symmetric pairs such as ARG-BRA and BRA-ARG in your data frame. Please proceed what you have for the sake of simplicity.
#If you can consolidate them, you can get 5 points bonus.)
for index, row in team_pairs.iterrows():
  x = row['Home Team Initials']
  y = row['Away Team Initials']

  for index2, row2 in team_pairs.iterrows():
    if row2['Home Team Initials'] == y and row2['Away Team Initials'] == x:
      team_pairs.loc[index, 'weight'] += row2['weight']
      team_pairs.loc[index, 'HomeGoalTotal'] += row2['HomeGoalTotal']

      team_pairs.drop(index2)
      break
```

```python
team_pairs.to_csv('team_pairs.csv', index=False)
```
